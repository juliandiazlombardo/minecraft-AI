```python
class AIBrain:
    def __init__(self, provider="openai"):
        self.provider = provider.lower()
        self.tools = ["tool0"]
        self.skills = []
        self.goal = ""
        self.plan = ""
        self.identity = "Eres un agente de IA jugando Minecraft."
        self.system_prompt = ""

    def set_tools(self, tools:list):
        self.tools += tools

    def set_skills(self, skills:list):
        self.skills += skills

    def set_goal(self, goal:str):
        self.goal = goal

    def set_plan(self, plan:str):
        self.plan = plan

    def set_identity(self, identity:str):
        self.identity = identity
        
    def update_system_prompt(self):
        self.system_prompt = f""" {self.identity} 
        Tienes acceso a Tools (funciones) y Skills (conocimiento) 
        tools: 
        {self.tools} 
        skills: 
        {self.skills} 
        Ayudate de ellos para cumplir tu meta 
        meta: 
        {self.goal} 
        para lograr tu meta debes seguir el siguiente plan: 
        {self.plan}
        """
```

```python
brain = AIBrain()
print(f"{brain.system_prompt}")
```

```python
brain.set_tools(["tool1","tool2","tool3"])
brain.update_system_prompt()
print(f"{brain.system_prompt}")
```
