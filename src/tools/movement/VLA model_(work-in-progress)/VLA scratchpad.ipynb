```python
index = json.load(open("data/all_10xx_Jun_29.json"))

SAMPLES = 50

base = index["basedir"]
paths = index["relpaths"]

os.makedirs("data/10xx/videos", exist_ok=True)
os.makedirs("data/10xx/actions", exist_ok=True)

start_time = time.perf_counter()
for p in paths[:SAMPLES]:
    for ext in [".mp4", ".jsonl"]:
        url = base + p.replace(".mp4", ext)
        
        if ext == ".mp4":
           fname = "data/10xx/videos/" + p.split("/")[-1].replace(".mp4", ext)
        elif ext == ".jsonl":
            fname = "data/10xx/actions/" + p.split("/")[-1].replace(".mp4", ext)
        if os.path.exists(fname):
            continue

        open(fname, "wb").write(requests.get(url).content)

end_time = time.perf_counter()

total_time = end_time - start_time

print(f"the download of {SAMPLES} samples took {total_time}")






```
